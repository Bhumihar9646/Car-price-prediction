```python
CompanyName = cars['CarName'].apply(lambda x : x.split(' ')[0])
cars.insert(3,"CompanyName",CompanyName)
cars.drop(['CarName'],axis=1,inplace=True)
cars.head()
```

```python
cars.CompanyName = cars.CompanyName.str.lower()
def replace_name(a,b):
    cars.CompanyName.replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')

cars.CompanyName.unique()
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)
plt.title('Car Price Distribution Plot')
sns.distplot(cars.price)

plt.subplot(1,2,2)
plt.title('Car Price Spread')
sns.boxplot(y=cars.price)

plt.show()
```

```python
plt.figure(figsize=(25, 6))

plt.subplot(1,3,1)
plt1 = cars.CompanyName.value_counts().plot(kind='bar',color="r")
plt.title('Companies Histogram')
plt1.set(xlabel = 'Car company', ylabel='Frequency of company')

plt.subplot(1,3,2)
plt1 = cars.fueltype.value_counts().plot(kind ='bar',color = "g")
plt.title('Fuel Type Histogram')
plt1.set(xlabel = 'Fuel Type', ylabel='Frequency of fuel type')

plt.subplot(1,3,3)
plt1 = cars.carbody.value_counts().plot(kind='bar', color = "olive")
plt.title('Car Type Histogram')
plt1.set(xlabel = 'Car Type', ylabel='Frequency of Car type')

plt.show()
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)
plt.title('Symboling Histogram')
sns.countplot(cars.symboling, palette=("cubehelix"))

plt.subplot(1,2,2)
plt.title('Symboling vs Price')
sns.boxplot(x=cars.symboling, y=cars.price, palette=("cubehelix"))

plt.show()
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)
plt.title('Engine Type Histogram')
sns.countplot(cars.enginesize, palette=("Blues"))

plt.subplot(1,2,2)
plt.title('Engine size vs Price')
sns.boxplot(x=cars.enginesize, y=cars.price, palette=("PuBuGn"))

plt.show()

df = pd.DataFrame(cars.groupby(['enginetype'])['price'].mean().sort_values(ascending = False))
df.plot.bar(figsize=(8,6))
plt.title('Engine Type vs Average Price')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets , linear_model
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# apply EDA on diabetes data set
# diabetes = datasets.load_diabetes()
diabetes=pd.read_csv("/content/diabetes.csv")
diabetes
```

```python
diabetes = diabetes.values
diabetes
```

```python
diabetes_x = diabetes[:,:8]
diabetes_x
```

```python
diabetes_y = diabetes[:,:8]
diabetes_y
```

```python
x_train , x_test, y_train, y_test= train_test_split(diabetes_x,diabetes_y,test_size=0.20,shuffle=False,random_state=45)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
model =linear_model.LinearRegression()
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pre = model.predict(x_test)
print(y_pre)
```

```python
print("mean squraed error is : ", mean_squared_error(y_test,y_pre))
```
